```python
%matplotlib inline
import csv
import numpy as np
import matplotlib.pyplot as plt
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
plt.rcParams.update({'axes.titlesize': 'large'})
plt.rcParams.update({'axes.labelsize': 'large'})
```

```python
short_channel_dc_data = []
# vds, vgs, freq, Ids (amps)
with open('short_channel_dc_sweep_data.csv', 'r') as f:
    reader = csv.reader(f)
    next(reader)
    for row in reader:
        row = [float(d) for d in row]
        short_channel_dc_data += [row]
```

```python
vds_vals = [0.1, 0.4, 0.9]
plt.figure()
for vds in vds_vals:
    data = [row for row in short_channel_dc_data if row[0] == round(vds,2)]
    vgs = [d[1] for d in data]
    ids = [d[3]*1000.0 for d in data]
    plt.plot(vgs, ids)
plt.ylabel('$I_{ds} (mA)$')
plt.xlabel('$V_{gs} (V)$')
plt.title('Short Channel MOSFET: $V_{gs}$ vs $I_{ds}$ for fixed $V_{ds}$')
plt.legend(['$V_{ds} = %s V$' % vds for vds in vds_vals], loc='upper left')
plt.savefig('lab0_report/images/short_channel_vgs_vs_ids.png', bbox_inches='tight', dpi=300)
plt.show()
```

```python
vgs_vals = [0.3, 0.45, 0.6, 0.9]
plt.figure()
for vgs in vgs_vals:
    data = [row for row in short_channel_dc_data if row[1] == round(vgs,2)]
    vds = [d[0] for d in data]
    ids = [d[3]*1000.0 for d in data]
    plt.plot(vds, ids)
plt.ylabel('$I_{ds} (mA)$')
plt.xlabel('$V_{ds} (V)$')
plt.title('Short Channel MOSFET: $V_{ds}$ vs $I_{ds}$ for fixed $V_{gs}$')
plt.legend(['$V_{gs} = %s V$' % vgs for vgs in vgs_vals], loc='upper left')
plt.savefig('lab0_report/images/short_channel_vds_vs_ids.png', bbox_inches='tight', dpi=300)
plt.show()
```

```python
short_channel_ac_gm_data = []
# vd_dc, vg_dc, freq, Ids (amps) (ac)
with open('short_channel_ac_gm.csv', 'r') as f:
    reader = csv.reader(f)
    next(reader)
    for row in reader:
        row = [float(d) for d in row]
        short_channel_ac_gm_data += [row]
```

```python
# gm = i_ds / v_in
# vin = 10mV
# first plot gm vs Vgs where Vds is fixed at 0.9V and freq is fixed at 100 Mhz
short_channel_ac_gm_data_vgs = [[row[1], row[3]/0.010] for row in short_channel_ac_gm_data if row[0] == 0.9 and row[2] == 10e6]
x = [row[0] for row in short_channel_ac_gm_data_vgs]
y = [row[1] for row in short_channel_ac_gm_data_vgs]
plt.figure()
plt.plot(x, y)
plt.ylabel('$g_m (Siemens)$')
plt.xlabel('$V_{gs} (V)$')
plt.title('Short Channel MOSFET: $g_m$ vs $V_{gs}$ for $V_{ds} = 0.9V$')
plt.savefig('lab0_report/images/short_channel_gm_vs_vgs.png', bbox_inches='tight', dpi=300)
plt.show()

short_channel_ac_gm_data_vds = [[row[0], row[3]/0.010] for row in short_channel_ac_gm_data if row[2] == 10e6 and row[1] == 0.5]
x = [row[0] for row in short_channel_ac_gm_data_vds]
y = [row[1] for row in short_channel_ac_gm_data_vds]
plt.figure()
plt.plot(x, y)
plt.ylabel('$g_m (Siemens)$')
plt.xlabel('$V_{ds} (V)$')
plt.title('Short Channel MOSFET: $g_m$ vs $V_{ds}$ for $V_{gs} = 0.5V$')
plt.savefig('lab0_report/images/short_channel_gm_vs_vds.png', bbox_inches='tight', dpi=300)
plt.show()
```

```python
short_channel_ac_ro_data = []
# vds_dc, ids_ac
with open('short_channel_ac_ro.csv', 'r') as f:
    reader = csv.reader(f)
    next(reader)
    for row in reader:
        row = [float(d) for d in row]
        short_channel_ac_ro_data += [row]
x = [row[0] for row in short_channel_ac_ro_data]
y = [0.010 / row[1] for row in short_channel_ac_ro_data]
plt.figure()
plt.plot(x, y)
plt.ylabel('$r_o (Ohms)$')
plt.xlabel('$V_{ds} (V)$')
plt.title('Short Channel MOSFET: $r_o$ vs $V_{ds}$ for $V_{gs} = 0.5V$')
plt.savefig('lab0_report/images/short_channel_ro_vs_vds.png', bbox_inches='tight', dpi=300)
plt.show()
```

```python
# voltage gain |A_V| = g_m * r_o
vds = [row[0] for row in short_channel_ac_ro_data]
ro = [0.010 / row[1] for row in short_channel_ac_ro_data]
gm = [row[3]/0.010 for row in short_channel_ac_gm_data if row[2] == 10e6 and row[1] == 0.5]
gain = np.multiply(ro, gm)
plt.figure()
plt.plot(vds, gain)
plt.ylabel('$A_V (Voltage Gain)$')
plt.xlabel('$V_{ds} (V)$')
plt.title('Short Channel MOSFET: $A_V$ vs $V_{ds}$ for $V_{gs} = 0.5V$')
plt.savefig('lab0_report/images/short_channel_av_vs_vds.png', bbox_inches='tight', dpi=300)
plt.show()
```

```python
long_channel_ac_gm_data = []
# vd_dc, vg_dc, freq, Ids (amps) (ac)
with open('long_channel_ac_gm.csv', 'r') as f:
    reader = csv.reader(f)
    next(reader)
    for row in reader:
        row = [float(d) for d in row]
        long_channel_ac_gm_data += [row]
```

```python
long_channel_ac_ro_data = []
# vds_dc, ids_ac
with open('long_channel_ac_ro.csv', 'r') as f:
    reader = csv.reader(f)
    next(reader)
    for row in reader:
        row = [float(d) for d in row]
        long_channel_ac_ro_data += [row]
x = [row[0] for row in long_channel_ac_ro_data]
y = [0.010 / row[1] for row in long_channel_ac_ro_data]
plt.figure()
plt.plot(x[0:30], y[0:30])
plt.ylabel('$r_o (Ohms)$')
plt.xlabel('$V_{ds} (V)$')
plt.title('Long Channel MOSFET: $r_o$ vs $V_{ds}$ for $V_{gs} = 3V$')
plt.savefig('lab0_report/images/long_channel_ro_vs_vds.png', bbox_inches='tight', dpi=300)
plt.show()
```

```python
# voltage gain |A_V| = g_m * r_o
vds = [row[0] for row in long_channel_ac_ro_data]
ro = [0.010 / row[1] for row in long_channel_ac_ro_data]
gm = [row[3]/0.010 for row in long_channel_ac_gm_data if row[2] == 10e6 and row[1] == 0.5]
gain = np.multiply(ro, gm)
plt.figure()
plt.plot(vds[0:31], np.multiply(20, np.log10(gain[0:31])))
plt.ylabel('$A_V (Voltage Gain, dB)$')
plt.xlabel('$V_{ds} (V)$')
plt.title('Long Channel MOSFET: $A_V$ vs $V_{ds}$ for $V_{gs} = 3V$')
plt.savefig('lab0_report/images/long_channel_av_vs_vds.png', bbox_inches='tight', dpi=300)
plt.show()
```
