```python
%matplotlib inline
import csv
import numpy as np
import matplotlib.pyplot as plt
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
plt.rcParams.update({'axes.titlesize': 'large'})
plt.rcParams.update({'axes.labelsize': 'large'})

def read_csv(filename, row_offset=3):
    data = []
    with open(filename, 'r') as f:
        reader = csv.reader(f)
        for i in range(0, row_offset):
            next(reader)
        for row in reader:
            row = [float(d) for d in row]
            data += [row]
    return data
```

```python
# format: frequency, dB
cable_phase_s11 = read_csv('data/CABLE_PHASE_S11.csv')
cable_phase_s21 = read_csv('data/CABLE_PHASE_S21.csv')
cable_phase_s21_phase = read_csv('data/CABLE_PHASE_S21_PHASE.csv')
cable_phase_s21_after_data_divide = read_csv('data/CABLE_PHASE_S21_AFTER_DATA_DIVIDE.csv')
cable_phase_s21_after_movement = read_csv('data/CABLE_PHASE_S21_AFTER_MOVEMENT.csv')
```

```python
plt.figure()
plt.plot([x[0] for x in cable_phase_s11], [x[1] for x in cable_phase_s11])
plt.xlabel('Frequency (Hz)')
plt.ylabel('Response (dB)')
plt.ylim([-90, 10])
plt.title('Cable Between Port 1/2, S11')
plt.savefig('images/cable_phase_s11.png', bbox_inches='tight', dpi=300)
plt.show()
```

```python
plt.figure()
plt.plot([x[0] for x in cable_phase_s21_phase], [x[1] for x in cable_phase_s21_phase])
plt.xlabel('Frequency (Hz)')
plt.ylabel('Phase (deg)')
plt.ylim([-220, 220])
plt.title('Cable Between Port 1/2, S21 (Phase)')
plt.savefig('images/cable_phase_s21_phase.png', bbox_inches='tight', dpi=300)
plt.show()
```

```python
plt.figure()
plt.plot([x[0] for x in cable_phase_s21_after_data_divide], [x[1] for x in cable_phase_s21_after_data_divide])
plt.xlabel('Frequency (Hz)')
plt.ylabel('Phase (deg)')
plt.ylim([-1, 1])
plt.title('Cable Between Port 1/2 After Divide, S21 (Phase)')
plt.savefig('images/cable_phase_s21_after_data_divide.png', bbox_inches='tight', dpi=300)
plt.show()
```

```python
plt.figure()
plt.plot([x[0] for x in cable_phase_s21_after_movement], [x[1] for x in cable_phase_s21_after_movement])
plt.xlabel('Frequency (Hz)')
plt.ylabel('Phase (deg)')
plt.ylim([-1, 1])
plt.title('Cable Between Port 1/2 After Divide and Movement, S21 (Phase)')
plt.savefig('images/cable_phase_s21_after_movement.png', bbox_inches='tight', dpi=300)
plt.show()
```
